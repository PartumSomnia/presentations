# Exercise 1
Source: https://nap.nationalacademies.org/jhp/oneuniverse/frontiers_exercise_8.html
> 1.1 How much energy do gamma-ray bursts produce?

The gamma-ray burst GRB031332 produces a flux of $8 \times 10^{-10}$ Watts per square
meter on Earth, for a period of about 90 seconds.  A followup measurement shows
that the burst occurred about 3000 Megaparsecs away.

(a) What was the luminosity of this gamma-ray burst?  How much energy did the
explosion produce during the $90$-second burst period?

(b) How does that luminosity compare to that of the entire Milky Way galaxy?
Remember that the Sun's luminosity is about $4 \times 10^{26}$ Watts.

(c) The energy released by a supernova explosion is about $10^{44}$ Joules.
About how many times more energy was released by this gamma-ray burst?

> 1.2 What could produce a burst of energy as powerful as GRB031332?

(a) If two neutron stars, each twice the mass of the Sun, fell toward
each other and collided, how much gravitational potential energy would 
be released?  How does this compare with the energy released by GRB031332?

(b) Would it be possible for a burst of energy of this size to be
released if ordinary stars or diffuse gas totaling 2 times the mass
of the Sun fell onto a 2-solar-mass neutron star?  Explain
your reasoning.

# Solutions
> 1.1 
(a) Since a Megaparsec is about 3.1 * 10^22 m, this burst's luminosity is
$$
   L = 4 \pi r^2 F = 4 \times 3.14 \times (3000 \times 3.1 \times 10^{22} \text{m})^2 \times (5 \times 10^{-10} \text{W/m}^2)
     = 8.7 \times 10^{43} \text{ W or } 8.7 \times 10^{43} \text{Joule/sec} \\
$$
   For a 90 second period, the total energy release was
$$
   E = (8.7 \times 10^{43} \text{Joule/s}) \times (90 \times{s}) = 7.8 \times 10^{45} \text{J}.
$$

(b) The luminosity of a galaxy is usually about equal to the luminosity of all
its stars put together.  Since the Milky Way has about $10^{11}$ stars, if we use
the Sun as a typical star then our galaxy's luminosity is
$$
   L = 10^{11} \times (4 \times 10^{26} \text{ W}) = 4 \times 10^{37} \text{ W}.
$$

The luminosity of the gamma-ray burst is 
$$
   \frac{8.7 \times 10^{43} \text{ W}}{4 \times 10^{37} \text{ W}} = 2.2
$$
So it is $2.2$ million times greater than an entire galaxy!

(c) The energy released by the burst was $(7.8 \times 10^{45})/(10^{44}) = 78$ times 
greater than that of a typical supernova.

> 1.2 
(a) The amount of energy released would be approximately equal to one neutron star
falling onto the surface of the other neutron star from a large distance.
The radius of a neutron star is about $10$ km.  If the other neutron star drops
from a distance of 1000 km, then the energy released would be
$$
   E = -GMm/r_{\rm start} - (-GMm/r_{\rm end}) \\
     = -(6.67 \times 10^{-11}) \times (4 \times 10^{30})^2 \times (1/(10^{6}) - 1/(10^{4})) \\
     = 1.1 \times 10^{47} \text{ J}.
$$
This is a little more than ten times the energy produced by GRB031332, so if
ten percent of the gravitational potential energy released were converted into
radiation, this neutron star collision scenario would be a plausible way to
create the gamma-ray burst.

(b) The total amount of gravitational potential energy released would be
the same if ordinary stars or diffuse gas fell onto a neutron star.  But
the timing would be very different; before that "ordinary" matter reached
the neutron star's surface, it would be pulled into an orbiting disk of
material, which would regulate a gradual flow of matter onto the neutron
star.  So instead of releasing the energy in a matter of seconds, it would
take years or even millennia ï¿½ not in the form of a burst.  Only a very
solid, tightly bound object like another neutron star could resist the
tidal disruption as it fell in, thus allowing the gravitational potential
energy to "burst" forth.
